```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("moon.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(6,6))
plt.imshow(img, cmap="gray")
plt.title("Original Moon Image")
plt.axis("off")
```

```python
laplacian_kernel = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
], dtype=np.float32)

print(laplacian_kernel)
```

```python
laplacian_img = cv2.filter2D(img, -1, laplacian_kernel)

plt.figure(figsize=(6,6))
plt.imshow(laplacian_img, cmap="gray")
plt.title("Laplacian Filter Result")
plt.axis("off")
```

```python
sharpened = cv2.addWeighted(img, 1, laplacian_img, 1, 0)

plt.figure(figsize=(6,6))
plt.imshow(sharpened, cmap="gray")
plt.title("Sharpened Image (Original + Laplacian)")
plt.axis("off")
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(laplacian_img, cmap="gray")
plt.title("Laplacian Output")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(sharpened, cmap="gray")
plt.title("Sharpened Image")
plt.axis("off")

plt.show()
```
